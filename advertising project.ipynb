```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#opening dataset
pd.pandas.set_option('display.max_columns',None)
dataset=pd.read_csv('advertising.csv')

print(dataset.shape)
    
dataset.head()
```

```python
#finding missing values
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#since there are no missing values, I will continue to the next step of exploratory data analysis
```

```python
dataset.info()
```

```python
dataset.plot(x='TV',y='sales',kind='line')
```

```python
#the above graph is ineffective! So, let's try a histogram.
```

```python
dataset[['TV','radio','newspaper']].hist(bins=15,figsize=(12,8))
plt.show()
```

```python
dataset[['TV','radio','newspaper']].plot.hist(bins=15,figsize=(12,8))

plt.show()
```

```python
dataset.corr()
```

```python
dataset.corr()['sales']
```

```python
#According to the correlation, the most effective channel is TV, then radio, then newspaper.
#let's see this data in a heatmap
sns.heatmap(dataset.corr(),annot=True, linewidth=0.5, linecolor='blue', fmt='.2f')
```

```python
dataset.skew()
```

```python
#skewness is highest for newspaper.

sns.distplot(dataset['TV'])
```

```python
sns.distplot(dataset['radio'])
```

```python
sns.distplot(dataset['newspaper'])
```

```python
#according to the above graphs, it can be understood that even though TV has less skewed data, radio has a more stable distribution and there is not much fluctuation, unlike TV. And newspaper is positively skewed.
```

```python
dataset.describe()
```

```python
#standard deviation is the highest for TV , which again shows the fluctuating property. This means that the values in TV are highly dispersed from the mean.
```

```python
sns.pairplot(dataset, hue='sales')
```
